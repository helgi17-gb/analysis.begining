### Задание 1

Даны 4 последовательности. Необходимо:

a) исследовать их на монотонность; 

b) исследовать на ограниченность;

c) найти пятый по счету член.


------------------

$$\{a_n\}_{n=1}^\infty = 2^n-n$$

$a_n - a_{n+1}=2^n-n - (2^{n+1}-(n+1))=2^n(1-2)-n+n+1=1-2^n$

Для $n\geq 1$ разность всегда отрицательна. Последовательность монотонно возрастает.

Последовательность неограничена. $\lim_{x\to +\infty}a_n=+\infty$

$$a_5=2^5-5=27$$

------------------

$$\{b_n\}_{n=2}^\infty = \frac{1}{1-n}$$

$\frac{b_n}{b_{n+1}}=\frac{1-(n+1)}{1-n}=\frac{(1-n)-1}{1-n}=1-\frac{1}{1-n}$

Для $n\geq 2$ дробь больше 1, то есть $b_n>b_{n+1}$. Так как, при этом, $b_n<0$, то последовательность монотонно возрастает.

Последовательность ограничена. $\lim_{n\to +\infty}b_n=0$

$$b_5=\frac{1}{1-5}=-0.25$$

------------------

$$\{c_n\}_{n=1}^\infty = -1^n+\sqrt{2n}$$

$c_n-c_{n+1}=-1^n+\sqrt{2n}-\left(-1^{n+1}+\sqrt{2n+2}\right)=-1^n(1-(-1))+\sqrt{2n}-\sqrt{2n+2}=-1^n*2+\sqrt{2n}-\sqrt{2n+2}$

Последовательность не является монотонной. При $n=1$ $c_n<c_{n+1}$, а при $n=2$ $c_n>c_{n+1}$.

Последовательность неограничена. $\lim_{n\to +\infty}c_n=+\infty$

$$c_5=-1^5+\sqrt{2*5}=-1+\sqrt{10}$$

------------------

$$\{d_n\}_{n=1}^\infty = \left(-1\right)^{2n}+\frac{1}{n^2}$$

$\frac{d_n}{d_{n+1}}=\frac{(-1)^{2n}+\frac{1}{n^2}}{(-1)^{2n+2}+\frac{1}{(n+1)^2}}=\frac{(n+1)^2}{n^2}$

Для $n\geq 1$ дробь больше 1. Последовательность монотонно убывает.

Последовательность ограничена. $\lim_{n\to +\infty}d_n=0$

$$d_5=(-1)^{2*5}+\frac{1}{5^2}=\frac{1}{25}=0.04$$

### Задание 3

*На языке Python предложить алгоритм вычисляющий численно предел с точностью 
$\epsilon=10^{-7}$

$$\lim_{n \to +\infty}\frac{n}{\sqrt[n]{n!}}$$

```python
import math

def nthroot_factorial_product(n):
    ret = 1.0
    for x in range(n):
        ret *= pow(x+1,1/n)
    return ret

def nthroot_factorial_stirling(n):
    ret = 1.0
    ret = (n/math.e)*pow(math.sqrt(2*math.pi*n),1/n)
    return ret

nthroot_factorial = nthroot_factorial_stirling
    


epsilon=10e-7
a_pred=1
n=2
a = n/nthroot_factorial(n)

while (abs(a-a_pred)>epsilon):
    print(n, a, a_pred)
    a_pred = a
    n += 1
    a = n/nthroot_factorial(n)
    
print(a)
```

Вычисление плохо сходится к $e$ при любом способе вычисления корня из факториала. Что странно.
